```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import nibabel as nib
```

```python
mypath = "../cyclegan_tf/data_train_MASK.npy"
maskfiles = np.load(mypath)

mypath2 = "../cyclegan_tf/filenames_train.npy"
filenames = np.load(mypath2)

mypath3 = "../cyclegan_tf/data_train_ADC.npy"
adcfiles = np.load(mypath3)
```

```python
#print(filenames.shape[0])
#print(maskfiles.shape)

lesion_areas = []
lesion_areas_index = []
num_lesion_slices = 0

for i in range(filenames.shape[0]):
    mysum = np.sum(maskfiles[i])
    print(i, filenames[i], mysum)
    lesion_areas.append(mysum)
    lesion_areas_index.append(i)
    num_lesion_slices += 1

print("")
print("num_lesion_slices:", num_lesion_slices, "... corresponds to: ", str(int(100*(num_lesion_slices/filenames.shape[0])))+"%")
print("biggest lesion:", np.max(lesion_areas), "... index[", lesion_areas_index[lesion_areas.index(np.max(lesion_areas))] ,"] ...corresponds to: ", filenames[lesion_areas_index[lesion_areas.index(np.max(lesion_areas))]])
print("smallest lesion:", np.min(lesion_areas), "... corresponds to: ", filenames[lesion_areas_index[lesion_areas.index(np.min(lesion_areas))]])
```
